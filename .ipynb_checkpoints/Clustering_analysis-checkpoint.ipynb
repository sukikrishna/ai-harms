```python
def combine_table(input_tables, patient_id_col):
  merged_data = input_tables[0]
  for i in range(1,len(input_tables)):
    merged_data = pd.merge(
            merged_data,
            input_tables[i],
            on=patient_id_col,
            how="inner",
            suffixes=('', str('_outcome' + str(i)))
        )
  return merged_data
```

## Misclassification scores

```python
def misclassification_score(data_table, actual_to_predicted_outcomes_dict, misclassification_colname):
    data_table[misclassification_colname] = 0
    colindex = 0
    for actual_outcome, predicted_outcomes in actual_to_predicted_outcomes_dict.items():
        # Loop through each predicted outcome for the current actual outcome
        for predicted_outcome in predicted_outcomes:
            comparison = (data_table[actual_outcome].fillna(-1) != data_table[predicted_outcome].fillna(-1))
            data_table[misclassification_colname] += comparison.astype(int)
            colindex += 1
    data_table[misclassification_colname] = (
        data_table[misclassification_colname] / colindex
    )

    return data_table
```

```python
def stratify_data(row, misclassification_colname):
    if row[misclassification_colname] == 0:
        return "No misclassification"
    elif 0 < row[misclassification_colname] < 0.5:
        return "Less than 50%"
    elif 0.5 <= row[misclassification_colname] <= 1:
        return "More than 50%"
    elif row[misclassification_colname] > 1:
        return "More than 1"
    else:
        return "Unknown"
```

```python
def plot_misclassification_histogram(data_table, misclassification_colname, bins=10):
    misclassification_scores = data_table[misclassification_colname]
    non_nan_scores = misclassification_scores.dropna()
    non_nan_scores = non_nan_scores[~non_nan_scores.isin([float('inf'), float('-inf')])]
    #nan_scores = misclassification_scores[misclassification_scores.isna()]

    plt.figure(figsize=(8, 6))
    plt.hist(non_nan_scores, bins=bins, edgecolor='black', alpha=0.7, label='Non-NaN')
    #plt.hist(nan_scores, bins=bins, edgecolor='black', alpha=0.7, label='NaN')
    plt.title(f'Histogram of {misclassification_colname}')
    plt.xlabel('Misclassification Score')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.legend(loc='upper right')
    plt.show()
```

```python
def stratification_summary(data, misclassification_colname, custom_order=None):
    #compute summary
    data[str('stratification_' + str(misclassification_colname))] = data.apply(lambda row: stratify_data(row, misclassification_colname), axis=1)
    summary = data.groupby(str('stratification_' + str(misclassification_colname)))['uniquepid'].count()
    print(summary)

    #Plot barplot
    if custom_order:
        summary = summary[custom_order]
    plt.figure(figsize=(8, 6))
    sns.barplot(x=summary.index, y=summary.values, palette='viridis')
    plt.title(str('Different categories of misclassified patients'), fontsize=14)
    plt.xlabel('Proportion of models for which the patient was misclassified', fontsize=12)
    plt.ylabel('Count of unique patients', fontsize=12)
    plt.xticks(rotation=45)
    plt.show()

    return data
```

```python
def plot_demographics_for_overall_and_subsets(df_features, combined_data, subset_definitions, all_uniquepids,
                                              categorical_columns=['ethnicity', 'gender'],
                                              numerical_columns=['age', 'weight']):
    df_features['gender'] = df_features['gender'].replace({0.0: 'men', 1.0: 'women'})
    sns.set_theme(style="whitegrid", palette="pastel")

    for subset_name, stratified_uniquepids in subset_definitions.items():
        subset_data = df_features[df_features['uniquepid'].isin(stratified_uniquepids)]
        full_data = df_features[df_features['uniquepid'].isin(all_uniquepids)]
        for column in categorical_columns:
            if column in df_features.columns:
                full_counts = df_features[column].value_counts(normalize=True)
                subset_counts = subset_data[column].value_counts(normalize=True)
                combined_counts = pd.DataFrame({
                    "Full Dataset": full_counts,
                    "Subset": subset_counts
                }).fillna(0)

                plt.figure(figsize=(8, 6))
                combined_counts.plot(kind='bar', width=0.8, color=['skyblue', 'salmon'], alpha=0.8)
                plt.title(f"Proportion of {column.capitalize()} ({subset_name} subset)", fontsize=14, fontweight='bold')
                plt.ylabel("Proportion", fontsize=12)
                plt.xlabel(column.capitalize(), fontsize=12)
                plt.xticks(rotation=45)
                plt.legend(title="Dataset")
                plt.grid(True, linestyle="--", alpha=0.6)
                plt.show()

        for column in numerical_columns:
            if column in df_features.columns:
                plt.figure(figsize=(8, 6))

                sns.kdeplot(
                    df_features[column].dropna(),
                    color="skyblue",
                    label="Full Dataset",
                    fill=False,
                    linewidth=2,
                )
                sns.kdeplot(
                    subset_data[column].dropna(),
                    color="salmon",
                    label="Subset",
                    fill=False,
                    linewidth=2,
                )

                plt.title(f"Density of {column.capitalize()} ({subset_name} subset)", fontsize=14, fontweight='bold')
                plt.xlabel(column.capitalize(), fontsize=12)
                plt.ylabel("Density", fontsize=12)
                plt.legend(title="Dataset")
                plt.grid(True, linestyle="--", alpha=0.6)
                plt.show()
```

```python
def plot_proportion_of_subset_with_error_bars(
    df_features,
    combined_data,
    subset_definitions,
    all_uniquepids,
    categorical_columns,
    confidence_level
):
    df_features['gender'] = df_features['gender'].replace({0.0: 'men', 1.0: 'women'})

    sns.set_theme(style="whitegrid", palette="pastel")

    z = stats.norm.ppf(1 - (1 - confidence_level) / 2)

    for subset_name, stratified_uniquepids in subset_definitions.items():
        subset_data = df_features[df_features['uniquepid'].isin(stratified_uniquepids)]
        full_data = df_features[df_features['uniquepid'].isin(all_uniquepids)]

        for column in categorical_columns:
            if column in df_features.columns:
                subset_counts = subset_data[column].value_counts(normalize=False)
                full_counts = full_data[column].value_counts(normalize=False)
                proportions = (subset_counts / full_counts).fillna(0)
                #errors = z * np.sqrt((proportions * (1 - proportions)) / full_counts).fillna(0)
                errors = standard_deviation = np.sqrt((proportions * (1 - proportions)) / full_counts).fillna(0)


                # Create a barplot with error bars
                plt.figure(figsize=(8, 6))
                sns.barplot(x=proportions.index, y=proportions.values, palette="pastel")
                plt.errorbar(
                    x=np.arange(len(proportions)),
                    y=proportions.values,
                    yerr=errors.values,
                    fmt='none',
                    c='black',
                    capsize=5
                )

                plt.title(f"Proportion of each {column} {subset_name}", fontsize=14, fontweight='bold')
                plt.ylabel("Proportion of Subset", fontsize=12)
                plt.xlabel(column.capitalize(), fontsize=12)
                plt.xticks(rotation=45)
                plt.grid(axis='y', linestyle="--", alpha=0.6)
                plt.show()
```

## Clustering

```python
def extract_misclassification_features(data, misclassified_pids):
    return data[data['uniquepid'].isin(misclassified_pids)]
```

```python
from sklearn.cluster import KMeans

def apply_kmeans(X_scaled, n_clusters=3, random_state=42):
    kmeans = KMeans(n_clusters=n_clusters, random_state=random_state)
    clusters = kmeans.fit_predict(X_scaled)
    return clusters
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

def visualize_clusters(X_scaled, clusters, n_components=2):
    pca = PCA(n_components=n_components)
    X_pca = pca.fit_transform(X_scaled)
    plt.figure(figsize=(8, 6))
    scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=clusters, cmap='viridis', alpha=0.7)
    plt.title("Clusters of Misclassified Patients", fontsize=14)
    plt.xlabel("PCA1", fontsize=12)
    plt.ylabel("PCA2", fontsize=12)
    plt.colorbar(scatter, label="Cluster")
    plt.grid(alpha=0.3)
    plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

def cluster_and_visualize(df_features_norm, misclassifiedpids, n_clusters=3):
    # Step 1: Filter the rows based on misclassified patients
    misclassified_data = df_features_norm[df_features_norm['uniquepid'].isin(misclassifiedpids)].drop(columns=['uniquepid'])

    # Step 2: Normalize the features if necessary
    scaler = StandardScaler()
    misclassified_data_scaled = scaler.fit_transform(misclassified_data)

    # Step 3: Apply KMeans clustering
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    clusters = kmeans.fit_predict(misclassified_data_scaled)

    # Add the cluster labels to the DataFrame
    df_features_norm.loc[df_features_norm['uniquepid'].isin(misclassifiedpids), 'cluster'] = clusters

    # Step 4: Visualize the clusters using PCA for dimensionality reduction
    pca = PCA(n_components=2)
    pca_components = pca.fit_transform(misclassified_data_scaled)

    plt.figure(figsize=(10, 8))
    plt.scatter(pca_components[:, 0], pca_components[:, 1], c=clusters, cmap='viridis', alpha=0.7)
    plt.title('Clusters of Misclassified Patients', fontsize=14)
    plt.xlabel('PCA1', fontsize=12)
    plt.ylabel('PCA2', fontsize=12)
    plt.colorbar(label='Cluster')
    plt.show()

    # Step 5: Extract the most influential features
    # The centroids of the clusters represent the "average" feature values for each cluster
    cluster_centroids = kmeans.cluster_centers_

    # Ensure that the number of features and cluster centroids align
    feature_importance = np.abs(cluster_centroids).mean(axis=0)

    # Check if the number of features and centroids match
    if len(feature_importance) != len(misclassified_data.columns):
        print(f"Warning: Mismatch in number of features ({len(misclassified_data.columns)}) and centroid values ({len(feature_importance)})")

    # Create a DataFrame to show feature importance
    feature_importance_df = pd.DataFrame({
        'feature': misclassified_data.columns,
        'importance': feature_importance
    }).sort_values(by='importance', ascending=False)

    # Show top influential features
    print("Top Influential Features:")
    print(feature_importance_df.head())

    return feature_importance_df
```

```python
def cluster_and_visualize_separation(df_features_norm, misclassifiedpids, n_clusters=3):
    # Step 1: Separate misclassified and correctly classified patients
    misclassified_data = df_features_norm[df_features_norm['uniquepid'].isin(misclassifiedpids)].drop(columns=['uniquepid'])
    non_misclassified_data = df_features_norm[~df_features_norm['uniquepid'].isin(misclassifiedpids)].drop(columns=['uniquepid'])

    # Combine both datasets for clustering
    combined_data = pd.concat([misclassified_data, non_misclassified_data], axis=0)

    # Step 2: Handle missing or infinite values
    # Replace NaN values with column means (you can also use median or drop rows/columns)
    combined_data = combined_data.fillna(combined_data.mean())

    # Step 3: Apply KMeans clustering on the entire dataset
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    combined_data['cluster'] = kmeans.fit_predict(combined_data)

    # Step 4: Visualize the clusters using PCA for dimensionality reduction
    pca = PCA(n_components=2)
    pca_components = pca.fit_transform(combined_data.drop(columns=['cluster']))  # Drop 'cluster' for PCA

    # Plot the scatter plot for all patients, coloring by cluster
    plt.figure(figsize=(10, 8))
    plt.scatter(pca_components[:, 0], pca_components[:, 1], c=combined_data['cluster'], cmap='viridis', alpha=0.7)
    plt.title('Clusters of All Patients (Misclassified vs Correctly Classified)', fontsize=14)
    plt.xlabel('PCA1', fontsize=12)
    plt.ylabel('PCA2', fontsize=12)
    plt.colorbar(label='Cluster')

    # Highlight misclassified patients in a different color
    misclassified_indices = df_features_norm['uniquepid'].isin(misclassifiedpids)
    plt.scatter(pca_components[misclassified_indices, 0], pca_components[misclassified_indices, 1], c='red', marker='x', label='Misclassified', alpha=1)

    plt.legend()
    plt.show()

    # Step 5: Extract the most influential features
    # Get cluster centroids (average feature values per cluster)
    cluster_centroids = kmeans.cluster_centers_

    # Calculate feature importance (absolute values of centroid feature means)
    feature_importance = np.abs(cluster_centroids).mean(axis=0)

    # Create a DataFrame to show feature importance (exclude 'cluster' from features)
    feature_importance_df = pd.DataFrame({
        'feature': combined_data.columns,  # All columns of features (including 'cluster' before it was dropped)
        'importance': feature_importance
    }).sort_values(by='importance', ascending=False)

    # Show top influential features
    print("Top Influential Features for Separation:")
    print(feature_importance_df.head())

    return feature_importance_df
```

```python
def cluster_and_select_separating_features(df_features_norm, misclassifiedpids, n_clusters=3):
    # Step 1: Separate misclassified and correctly classified patients
    misclassified_data = df_features_norm[df_features_norm['uniquepid'].isin(misclassifiedpids)].drop(columns=['uniquepid'])
    non_misclassified_data = df_features_norm[~df_features_norm['uniquepid'].isin(misclassifiedpids)].drop(columns=['uniquepid'])

    # Combine both datasets for clustering
    combined_data = pd.concat([misclassified_data, non_misclassified_data], axis=0)

    # Step 2: Handle missing or infinite values
    combined_data = combined_data.fillna(combined_data.mean())

    # Step 3: Apply KMeans clustering
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    combined_data['cluster'] = kmeans.fit_predict(combined_data)

    # Step 4: Create a binary label for misclassified vs correctly classified patients
    combined_data['is_misclassified'] = combined_data['uniquepid'].isin(misclassifiedpids).astype(int)

    # Step 5: Calculate the importance of features
    feature_importance = []

    for feature in combined_data.columns[:-2]:  # Exclude 'cluster' and 'is_misclassified'
        # Compare feature distributions between misclassified and correctly classified patients
        misclassified_values = combined_data[combined_data['is_misclassified'] == 1][feature]
        correctly_classified_values = combined_data[combined_data['is_misclassified'] == 0][feature]

        # Calculate mean difference between the two groups
        mean_diff = np.abs(misclassified_values.mean() - correctly_classified_values.mean())
        feature_importance.append((feature, mean_diff))

    # Step 6: Sort features by the mean difference (highest first)
    feature_importance_df = pd.DataFrame(feature_importance, columns=['feature', 'mean_difference'])
    feature_importance_df = feature_importance_df.sort_values(by='mean_difference', ascending=False)

    # Step 7: Visualize the features that best separate misclassified from correctly classified patients
    print("Top Features that Separate Misclassified and Correctly Classified Patients:")
    print(feature_importance_df.head())

    return feature_importance_df
```

## SVM

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay


def train_svm_and_get_importance_and_performance(df_features_norm, misclassifiedpids):
    # Step 1: Label the data
    df_features_norm['label'] = df_features_norm['uniquepid'].isin(misclassifiedpids).astype(int)

    # Separate the features (X) and target (y)
    X = df_features_norm.drop(columns=['uniquepid', 'label', 'cluster'])  # Drop 'uniquepid' and 'label' columns
    y = df_features_norm['label']

    # Step 2: Handle missing values - fill NaN values with the column mean
    X = X.fillna(X.mean())  # Fill NaN values with the mean of each column

    # Step 4: Train a linear SVM classifier
    svm = SVC(kernel='linear', random_state=42, class_weight='balanced')
    svm.fit(X, y)

    # Step 5: Get feature importance (coefficients)
    feature_importance = np.abs(svm.coef_.flatten())  # Absolute values of coefficients

    # Create a DataFrame to show feature importance
    importance_df = pd.DataFrame({
        'feature': X.columns,
        'importance': feature_importance
    }).sort_values(by='importance', ascending=False)

    # Output the top influential features
    print("Top Influential Features for Separating Misclassified Patients:")
    print(importance_df.head())

    # Step 6: Model performance evaluation
    y_pred = svm.predict(X)  # Predict labels using the trained SVM

    # Print performance metrics
    print("\nClassification Report:")
    print(classification_report(y, y_pred))

    # Plot Confusion Matrix
    cm = confusion_matrix(y, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot(cmap='Blues')
    plt.title('Confusion Matrix of SVM Classifier')
    plt.show()

    return importance_df
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import export_text

def train_random_forest_and_get_importance_and_performance(df_features_norm, misclassifiedpids):
    # Step 1: Label the data
    df_features_norm['label'] = df_features_norm['uniquepid'].isin(misclassifiedpids).astype(int)

    # Separate the features (X) and target (y)
    X = df_features_norm.drop(columns=['uniquepid', 'label', 'cluster'])  # Drop 'uniquepid', 'label', and 'cluster'
    y = df_features_norm['label']

    # Step 2: Handle missing values - fill NaN values with the column mean
    X = X.fillna(X.mean())

    # Step 3: Train a Random Forest classifier
    rf = RandomForestClassifier(
        n_estimators=100,        # Number of trees
        random_state=42,         # For reproducibility
        class_weight='balanced', # Adjust weights for imbalance
        max_depth=10             # Limit depth to avoid overfitting (can be tuned)
    )
    rf.fit(X, y)

    # Step 4: Get feature importance
    feature_importance = rf.feature_importances_  # Extract feature importance
    importance_df = pd.DataFrame({
        'feature': X.columns,
        'importance': feature_importance
    }).sort_values(by='importance', ascending=False)

    # Output the top influential features
    print("Top Influential Features for Separating Misclassified Patients:")
    print(importance_df.head())

    # Step 5: Model performance evaluation
    y_pred = rf.predict(X)  # Predict labels using the trained Random Forest

    # Print performance metrics
    print("\nClassification Report:")
    print(classification_report(y, y_pred))

    # Plot Confusion Matrix
    cm = confusion_matrix(y, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot(cmap='Blues')
    plt.title('Confusion Matrix of Random Forest Classifier')
    plt.show()

    return rf, X, importance_df
```
